# Smart Lender — Exploratory Data Analysis & Model Comparison

This notebook explores the loan applicant dataset and compares the four classification models used in the Smart Lender project: Decision Tree, Random Forest, KNN, and XGBoost.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json

sns.set_style('whitegrid')
df = pd.read_csv('../data/loan_data.csv')
df.head()
```

```python
df.info()
df.isnull().sum()
```

## Target distribution

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Loan_Status', palette=['#b4472d', '#2e7d6b'])
plt.title('Loan Status Distribution')
plt.show()
```

## Credit history vs. loan approval

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Credit_History', hue='Loan_Status')
plt.title('Credit History vs Loan Status')
plt.show()
```

## Applicant income distribution by loan status

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Loan_Status', y='ApplicantIncome')
plt.ylim(0, df['ApplicantIncome'].quantile(0.95))
plt.title('Applicant Income by Loan Status')
plt.show()
```

## Correlation heatmap (numeric features)

```python
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
plt.figure(figsize=(6,5))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='RdYlGn', center=0)
plt.title('Feature Correlation')
plt.show()
```

## Model comparison

Loads the metrics produced by `train_model.py` (run that script first so `model/model_comparison.json` exists).

```python
with open('../model/model_comparison.json') as f:
    comparison = json.load(f)

results_df = pd.DataFrame(comparison['results']).T
results_df
```

```python
ax = results_df.plot(kind='bar', figsize=(8,5), color=['#c99a3e', '#101a2b'])
plt.title('Model Comparison — Train vs Test Accuracy')
plt.ylabel('Accuracy (%)')
plt.xticks(rotation=20, ha='right')
plt.ylim(0, 100)
plt.legend(['Train Accuracy', 'Test Accuracy'])
plt.tight_layout()
plt.show()

print(f"Best performing model: {comparison['best_model']}")
```
